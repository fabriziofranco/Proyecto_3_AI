```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
df= pd.read_csv("experimentos/prueba.csv", header=None, names = ['Train_error','Validation_error'])
plt.plot(df.index, df['Train_error'])
plt.plot(df.index, df['Validation_error'])
plt.xlabel('Iteriaciones')
plt.ylabel('Cross-entropy')
plt.legend(['Train_error','Validation_error'])
```

```python
df= pd.read_csv("experimentos/tanh_3_capas_normal/error_and_accuracy.csv", header=None, names = ['Train_error','Validation_error', 'Train_acc','Validation_acc'])

fig, ax= plt.subplots(1, 2, figsize=(10,5))
ax[0].set_ylim([0,1.25])
ax[1].set_ylim([0,1])

ax[0].plot(df.index/1000, df['Train_error'], c='r')
ax[0].plot(df.index/1000, df['Validation_error'], c= 'g')
ax[0].set_xlabel('Iteraciones en miles')
ax[0].set_ylabel('Cross-entropy')


ax[1].plot(df.index/1000, df['Train_acc'], c='r')
ax[1].plot(df.index/1000, df['Validation_acc'],c='g')
ax[1].set_xlabel('Iteraciones en miles')
ax[1].set_ylabel('Accuracy')


ax[0].legend(['Train_error','Validation_error'],loc='lower left')
ax[1].legend(['Train_acc','Validation_acc'],loc='upper left')
```
